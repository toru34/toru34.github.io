```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 - 2

f_grad = jax.grad(f)
```

```python
ax = plt.subplot()

x = jnp.linspace(0, 3, 100)
y = f(x)

ax.plot(x, y)

x_tm1 = 0.4
ax.scatter(x_tm1, f(x_tm1))

for i in range(3):
    x_t = x_tm1 - f(x_tm1) / f_grad(x_tm1)
    
    ax.scatter(x_t, f(x_t), color='red')
    ax.text(x_t+0.1, f(x_t)-0.1, s=r'$x_' + str(i) + '$')
    
    x_tm1 = x_t

plt.show()
```
