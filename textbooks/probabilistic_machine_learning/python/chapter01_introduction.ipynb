```python
import pandas as pd
import altair as alt
from sklearn import datasets
```

```python
iris = datasets.load_iris()

df_feature = pd.DataFrame(iris.data, columns=iris.feature_names)
df_target = pd.DataFrame([iris.target_names[t] for t in iris.target], columns=['type'])

df = pd.concat([df_feature, df_target], axis=1)
df.head()
```

```python
alt.Chart(df).mark_point().encode(
    x='sepal length (cm)',
    y='sepal width (cm)',
    color='type',
).interactive()
```
