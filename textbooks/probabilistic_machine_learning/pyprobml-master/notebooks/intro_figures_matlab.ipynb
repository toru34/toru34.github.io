```

import cv2
from google.colab.patches import cv2_imshow
# Helper code to display images
def display_image(image,ratio):
    img = cv2.imread(image, cv2.IMREAD_UNCHANGED)
    img=cv2.resize(img, (0,0), fx=ratio, fy=ratio) 
    cv2_imshow(img)
    print("\n")
```
